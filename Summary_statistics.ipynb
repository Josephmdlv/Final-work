```python
Corruption_Index_2003_2020 = pandas.read_csv('Corruption_Index_2003_2020.csv')
Corruption_Index_2003_2020
```

```python
Corruption_Index_2003_2020 = Corruption_Index_2003_2020.dropna(axis=1, how="all")
if Corruption_Index_2003_2020["Year"].notna().all():
    Corruption_Index_2003_2020["Year"] = Corruption_Index_2003_2020["Year"].astype(int)
Corruption_Index_2003_2020 = Corruption_Index_2003_2020.iloc[:-1]
Corruption_Index_2003_2020["Year"] = Corruption_Index_2003_2020["Year"].astype(int)
assert Corruption_Index_2003_2020["Score"].between(0, 100).all(), "Invalid Score values found!"
Corruption_Index_2003_2020 = Corruption_Index_2003_2020.drop_duplicates()
Corruption_Index_2003_2020
```

```python
Corruption_Index_2003_2020 = Corruption_Index_2003_2020.pivot(index="Year", columns="Country", values="Score")
Corruption_Index_2003_2020 = Corruption_Index_2003_2020.sort_index().sort_index(axis=1)
Corruption_Index_2003_2020['Kuwait'] = Corruption_Index_2003_2020['Kuwait'].combine_first(Corruption_Index_2003_2020['Kuweit'])
Corruption_Index_2003_2020 = Corruption_Index_2003_2020.drop(columns=['Kuweit'])
Corruption_Index_2003_2020['Palestine'] = Corruption_Index_2003_2020['Palestine'].combine_first(Corruption_Index_2003_2020['Palestinian Authority'])
Corruption_Index_2003_2020 = Corruption_Index_2003_2020.drop(columns=['Palestinian Authority'])
Corruption_Index_2003_2020
```

```python
Corruption_Index_2003_2020.to_csv('Corruption_Index_2003_2020_Cleaned.csv')
```

```python
Summary_Stats_Corruption = Corruption_Index_2003_2020.describe()
Summary_Stats_Corruption
```

```python
Summary_Stats_Corruption.to_csv('Summary_Stats_Corruption.csv')
```

```python
human_development_index = pandas.read_csv('human-development-index.csv')
human_development_index
```

```python
human_development_index = human_development_index.rename(columns={"Entity": "Country"})
human_development_index = human_development_index.rename(columns={"Human Development Index": "HDI"})
human_development_index = human_development_index.drop(columns=["Code"])
human_development_index
```

```python
human_development_index = human_development_index.dropna(axis=1, how="all")
if human_development_index["Year"].notna().all():
    human_development_index["Year"] = human_development_index["Year"].astype(int)
human_development_index
```

```python
human_development_index["Year"] = human_development_index["Year"].astype(int)
```

```python
assert human_development_index["HDI"].between(0, 1).all(), "Invalid Score values found!"
```

```python
human_development_index = human_development_index.drop_duplicates()
```

```python
human_development_index = human_development_index.pivot(index="Year", columns="Country", values="HDI")
human_development_index = human_development_index.sort_index().sort_index(axis=1)
display (human_development_index)
```

```python
sum_HDI = human_development_index.describe()
display (sum_HDI)
sum_HDI.to_csv('sum_HDI.csv', index=False)
```
