```python
Corruption_Index_2003_2020 = pandas.read_csv('Corruption_Index_2003_2020.csv')
Corruption_Index_2003_2020
```

```python
corruption_df_cleaned = Corruption_Index_2003_2020.dropna(axis=1, how="all")
if corruption_df_cleaned["Year"].notna().all():
    corruption_df_cleaned["Year"] = corruption_df_cleaned["Year"].astype(int)
corruption_df_cleaned = corruption_df_cleaned.iloc[:-1]
corruption_df_cleaned
```

```python
corruption_df_cleaned["Year"] = corruption_df_cleaned["Year"].astype(int)
```

```python
assert corruption_df_cleaned["Score"].between(0, 100).all(), "Invalid Score values found!"
```

```python
corruption_df_cleaned = corruption_df_cleaned.drop_duplicates()
```

```python
reshaped_df = corruption_df_cleaned.pivot(index="Year", columns="Country", values="Score")
reshaped_df = reshaped_df.sort_index().sort_index(axis=1)
display (reshaped_df)
```

```python
reshaped_df.to_csv('corruption_df_cleaned.csv')
```

```python
summary_stats_corruption = reshaped_df.describe()
display (summary_stats_corruption)
```

```python
summary_stats_corruption.to_csv('summary_stats_corruption.csv')
```
